```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
#Check the data
purchase_data.head()
```

```python
#Find all the unique players
all_players = purchase_data["SN"].value_counts()
all_unique_players = all_players.count()
```

```python
#Put it into a dataframe
number_of_players_data = {"Total Players": [all_unique_players]}
total_players = pd.DataFrame(data = number_of_players_data)
total_players
```

```python
#Get avg price
avg_price = round(purchase_data["Price"].mean(), 2)

#Get number of purchases
number_purchases = purchase_data["Purchase ID"].count()

#Get total revenue
total_revenue = round(purchase_data["Price"].sum(), 2)

#Get number of unique items
unique_item_data = purchase_data["Item ID"].value_counts()
unique_item = unique_item_data.count()
```

```python
#Create the dataframe
purchasing_analysis = pd.DataFrame({"Number of Unique Items": [unique_item], "Average Price": [avg_price],
                       "Number of Purchases": [number_purchases], "Total Revenue": [total_revenue]})

#Format the dataframe
purchasing_analysis["Average Price"] = purchasing_analysis["Average Price"].astype(float).map("${:,.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)

#Display dataframe
purchasing_analysis
```

## Gender Demographics

```python
#Male Data
purchase_data_male = purchase_data.loc[purchase_data["Gender"] == "Male", :]
grouped_data_male = purchase_data_male.groupby("SN")
male_total = grouped_data_male["Gender"].count()
all_male = male_total.count()

percent_male = (all_male/all_unique_players)*100

#Female Data
purchase_data_female = purchase_data.loc[purchase_data["Gender"] == "Female", :]
grouped_data_female = purchase_data_female.groupby("SN")
female_total = grouped_data_female["Gender"].sum()
all_female = female_total.count()

percent_female = (all_female/all_unique_players)*100

#Other Data
all_other = (all_unique_players - (all_male + all_female))

percent_other = (all_other/all_unique_players)*100
```

```python
#Create the dataframe
gender_demographics = pd.DataFrame({"Gender": ["Male","Female","Other/Non-Disclosure"],
                                   "Total Count": [all_male, all_female, all_other],
                                   "Percentage of Players": [percent_male, percent_female, percent_other]})

#Format the dataframe
gender_demographics["Percentage of Players"] = gender_demographics["Percentage of Players"].astype(float).map("{:,.2f}".format)

#Reset the dataframe index
gender_demographics.set_index("Gender")
```


## Purchasing Analysis (Gender)

```python
#Get the purchase count
purchase_count = purchase_data.groupby("Gender")["Item ID"].count()
purchase_count_female = purchase_count["Female"]
purchase_count_male = purchase_count["Male"]
purchase_count_other = purchase_count["Other / Non-Disclosed"]

#Get the avg purchase price
avg_purchase_price = purchase_data.groupby("Gender")["Price"].mean()
avg_purchase_price_female = avg_purchase_price["Female"]
avg_purchase_price_male = avg_purchase_price["Male"]
avg_purchase_price_other = avg_purchase_price["Other / Non-Disclosed"]


#Get the total purchase price
sum_purchase_price = purchase_data.groupby("Gender")["Price"].sum()
sum_purchase_price_female = sum_purchase_price["Female"]
sum_purchase_price_male = sum_purchase_price["Male"]
sum_purchase_price_other = sum_purchase_price["Other / Non-Disclosed"]

#Break it down by gender
avg_female = (sum_purchase_price["Female"]/all_female)
avg_male = (sum_purchase_price["Male"]/all_male)
avg_other = (sum_purchase_price["Other / Non-Disclosed"]/all_other)
```

```python
#Create the dataframe
purchase_analysis_gender = pd.DataFrame({"Gender": ["Female","Male","Other/Non-Disclosed"],
                                        "Purchase Count": [purchase_count_female, purchase_count_male, purchase_count_other],
                                        "Average Purchase Price": [avg_purchase_price_female, avg_purchase_price_male, avg_purchase_price_other],
                                        "Total Purchase Value": [sum_purchase_price_female, sum_purchase_price_male, sum_purchase_price_other],
                                        "Avg Total Purchase per Person": [avg_female, avg_male, avg_other]})

#Format the dataframe
purchase_analysis_gender["Average Purchase Price"] = purchase_analysis_gender["Average Purchase Price"].astype(float).map("${:,.2f}".format)
purchase_analysis_gender["Total Purchase Value"] = purchase_analysis_gender["Total Purchase Value"].astype(float).map("${:,.2f}".format)
purchase_analysis_gender["Avg Total Purchase per Person"] = purchase_analysis_gender["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

#Reset the dataframe index
purchase_analysis_gender.set_index("Gender")
```

## Top Spenders

```python
#Get total purchase count
purchase_count = purchase_data.groupby("SN")["Item ID"].count()
purchase_count_df = pd.DataFrame(data = purchase_count)
purchase_count_df2 = purchase_count_df.rename(columns={'SN': 'SN', 'Item ID': 'Total Count'})

#Get total purchase value
total_purchase_value = purchase_data.groupby("SN")["Price"].sum()
total_purchase_value_df = pd.DataFrame(data = total_purchase_value)
total_purchase_value_df2 = total_purchase_value_df.rename(columns={'SN': 'SN', 'Price': 'Total Purchase Value'})

#Get avg purchase price
avg_purchase_price = (total_purchase_value/purchase_count)
avg_purchase_price_df = pd.DataFrame(data = avg_purchase_price)
avg_purchase_price_df2 = avg_purchase_price_df.rename(columns={'SN': 'SN', 0: 'Average Purchase Price'})

#Merge purchase count and avg price
merge_table = pd.merge(purchase_count_df2, avg_purchase_price_df2, on="SN")

#merge 1st merge and total value
merge_table_2 = pd.merge(merge_table, total_purchase_value_df2, on="SN")

#sort values by top total spenders
merge_table_3 = merge_table_2.sort_values("Total Purchase Value",ascending=False)

merge_table_3["Average Purchase Price"] = merge_table_3["Average Purchase Price"].astype(float).map("${:,.2f}".format)
merge_table_3["Total Purchase Value"] = merge_table_3["Total Purchase Value"].astype(float).map("${:,.2f}".format)

merge_table_3.head()
```

## Most Popular Items

```python
#Get Total purchase count 
pop_purchase_count = purchase_data.groupby("Item ID")["Item Name"].value_counts()
pop_purchase_count_df = pd.DataFrame(pop_purchase_count)
pop_purchase_count_df2 = pop_purchase_count_df.rename(columns={'Item Name': 'Purchase Count'})

#Get Total Item Price
total_pop_item_price = purchase_data.groupby("Item ID")["Price"].sum()
total_pop_item_price_df = pd.DataFrame(total_pop_item_price)
total_pop_item_price_df2 = total_pop_item_price_df.rename(columns={'Price': 'Total Purchase Value'})

#Get Item Price
pop_item_price = (total_pop_item_price/pop_purchase_count)
pop_item_price_df = pd.DataFrame(data = pop_item_price)
pop_item_price_df2 = pop_item_price_df.rename(columns={0: 'Item Price'})

#Merge Table 1
pop_merge_table = pd.merge(pop_purchase_count_df2, pop_item_price_df2, on=["Item ID","Item Name"])

#Merge Table 2
pop_merge_table_2 = pop_merge_table.join(total_pop_item_price_df2, on="Item ID")

#Sort Table 2
pop_merge_table_3 = pop_merge_table_2.sort_values("Purchase Count",ascending=False)

#Format new dataframe
pop_merge_table_3["Item Price"] = pop_merge_table_3["Item Price"].astype(float).map("${:,.2f}".format)
pop_merge_table_3["Total Purchase Value"] = pop_merge_table_3["Total Purchase Value"].astype(float).map("${:,.2f}".format)

pop_merge_table_3.head()
```
